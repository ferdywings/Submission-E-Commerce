## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa kategori produk yang paling banyak terjual?


```python
# Menggabungkan order_items dengan produk untuk mendapatkan nama kategori
merged_df = order_items_df.merge(products_df, on="product_id", how="left")

# Menggabungkan dengan terjemahan kategori untuk mendapatkan nama dalam bahasa Inggris
merged_df = merged_df.merge(product_category_df, on="product_category_name", how="left")

# Menghitung jumlah produk yang terjual per kategori
category_sales = merged_df["product_category_name_english"].value_counts().reset_index()
category_sales.columns = ["Kategori", "Jumlah Penjualan"]

# Mengatur ukuran gambar
plt.figure(figsize=(12, 6))

# Membuat diagram batang
sns.barplot(data=category_sales.head(10), x="Jumlah Penjualan", y="Kategori", hue="Kategori", palette="viridis", legend=False)

# Menambahkan judul dan label
plt.xlabel("Jumlah Penjualan", fontsize=12)
plt.ylabel("Kategori Produk", fontsize=12)
plt.title("10 Kategori Produk Terlaris", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Menampilkan plot
plt.show()
```

**Insight:**
- Kategori "Bed, Bath & Table" memiliki jumlah penjualan tertinggi, menunjukkan bahwa produk rumah tangga sangat diminati.
- Kategori "Health & Beauty" dan "Sports & Leisure" juga memiliki penjualan tinggi, mengindikasikan bahwa pelanggan memiliki minat besar pada produk perawatan diri dan olahraga.
- Kategori "Computers & Accessories" serta "Furniture & Decor" juga termasuk dalam daftar teratas, menunjukkan bahwa banyak pelanggan membeli produk teknologi dan dekorasi rumah.

###ClusteringAnalysis

`Analisis Clustering Berdasarkan Kategori dan Harga Produk`

```python
# Load datasets
order_items = pd.read_csv("/content/order_items_dataset.csv")
products = pd.read_csv("/content/products_dataset.csv")

# Merge datasets untuk mendapatkan kategori produk dan harga
merged_df = order_items.merge(products, on="product_id", how="left")

# Manual Grouping berdasarkan kategori produk
category_groups = merged_df.groupby("product_category_name").size().reset_index(name="count")

# Binning harga produk
price_bins = [0, 50, 200, 500, 1000, float("inf")]
price_labels = ["Sangat Murah", "Murah", "Sedang", "Mahal", "Sangat Mahal"]
merged_df["price_category"] = pd.cut(merged_df["price"], bins=price_bins, labels=price_labels, include_lowest=True)

# Hasil binning harga produk
price_groups = merged_df.groupby("price_category", observed=False).size().reset_index(name="count")

# Print hasil clustering
print("Clustering berdasarkan kategori produk:")
print(category_groups)
print("\nClustering berdasarkan harga produk:")
print(price_groups)
```

`Analisis Clustering Produk dan Transaksi`

```python
order_items = pd.read_csv("/content/order_items_dataset.csv")
products = pd.read_csv("/content/products_dataset.csv")
order_reviews = pd.read_csv("/content/order_reviews_dataset.csv")
order_payments = pd.read_csv("/content/order_payments_dataset.csv")

# Merge datasets untuk mendapatkan kategori produk, harga, ulasan, dan pembayaran
merged_df = order_items.merge(products, on="product_id", how="left")
merged_df = merged_df.merge(order_reviews, on="order_id", how="left")
merged_df = merged_df.merge(order_payments, on="order_id", how="left")

# Manual Grouping berdasarkan kategori produk
category_groups = merged_df.groupby("product_category_name").size().reset_index(name="count")

# Binning harga produk
price_bins = [0, 50, 200, 500, 1000, float("inf")]
price_labels = ["Sangat Murah", "Murah", "Sedang", "Mahal", "Sangat Mahal"]
merged_df["price_category"] = pd.cut(merged_df["price"], bins=price_bins, labels=price_labels, include_lowest=True)
price_groups = merged_df.groupby("price_category", observed=False).size().reset_index(name="count")

# Clustering berdasarkan ulasan pelanggan
review_bins = [1, 2, 3, 5]
review_labels = ["Buruk", "Sedang", "Baik"]
merged_df["review_category"] = pd.cut(merged_df["review_score"], bins=review_bins, labels=review_labels, include_lowest=True)
review_groups = merged_df.groupby("review_category", observed=False).size().reset_index(name="count")

# Clustering berdasarkan metode pembayaran
payment_groups = merged_df.groupby("payment_type").size().reset_index(name="count")

# Binning berdasarkan jumlah cicilan
installment_bins = [0, 1, 5, float("inf")]
installment_labels = ["Tunai", "Cicilan Pendek", "Cicilan Panjang"]
merged_df["installment_category"] = pd.cut(merged_df["payment_installments"], bins=installment_bins, labels=installment_labels, include_lowest=True)
installment_groups = merged_df.groupby("installment_category", observed=False).size().reset_index(name="count")

# Binning berdasarkan berat produk
weight_bins = [0, 500, 2000, float("inf")]
weight_labels = ["Ringan", "Sedang", "Berat"]
merged_df["weight_category"] = pd.cut(merged_df["product_weight_g"], bins=weight_bins, labels=weight_labels, include_lowest=True)
weight_groups = merged_df.groupby("weight_category", observed=False).size().reset_index(name="count")

# Print hasil clustering
print("Clustering berdasarkan kategori produk:")
print(category_groups)
print("\nClustering berdasarkan harga produk:")
print(price_groups)
print("\nClustering berdasarkan ulasan pelanggan:")
print(review_groups)
print("\nClustering berdasarkan metode pembayaran:")
print(payment_groups)
print("\nClustering berdasarkan jumlah cicilan:")
print(installment_groups)
print("\nClustering berdasarkan berat produk:")
print(weight_groups)
```
